```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import cm
from PIL import Image
import requests

img=cv2.imread("../Resources/taiwan_mountain.png")


gimg=img[:,:,1]   
gimg=gimg.reshape(-1)

sli=int(len(gimg)/10)
lis=[]
for t in range(0,10):
    lis+=[sum(gimg[t*sli:(t+1)*sli])]


#Normalization
min_val = min(lis)  #find min
max_val = max(lis)  #find max
normalized_data = [(x - min_val) / (max_val - min_val) for x in lis] #calculate
print(normalized_data)
```
